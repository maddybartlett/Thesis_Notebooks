```python
import numpy as np
import pandas as pd
import random
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pickle
```

# Examine PCA Models

```python
#seed = 0
df1=pd.DataFrame()
for seed in range(20):
    pickle_filename = "Training_Out/%s_pca_model.pkl" % seed
    with open(pickle_filename, 'rb') as file:
        pca_model = pickle.load(file)

    loadings = pca_model.components_

    df1=df1.append(pd.DataFrame(loadings))
    
plt.scatter(*loadings, alpha=0.3, label="Loadings");

plt.title("Loading plot");
plt.xlabel("Loadings on PC1");
plt.ylabel("Loadings on PC2");
plt.grid();
plt.legend(loc='lower left');
```

```python
export_csv = df1.to_csv ('pca_loadings.csv', index = None, header=True)
```
