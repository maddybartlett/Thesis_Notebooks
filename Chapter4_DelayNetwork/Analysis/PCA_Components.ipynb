```python
import numpy as np
import pandas as pd
import random
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pickle
```

# Examine PCA Models

## Variance Explained

```python
high = np.load('..\Data\Input\higheng.npy', allow_pickle=True)
low = np.load('..\\Data\\Input\\loweng.npy', allow_pickle=True)
```

```python
high_train_list=list(high[:,1])
low_train_list=list(low[:,1])
```

```python
train_all = np.vstack(high_train_list+low_train_list)
```

```python
pca_model = PCA(n_components=4).fit(train_all)
```

```python
print(pca_model.explained_variance_ratio_)
```

## Factor Loadings 

```python
#seed = 0
df1=pd.DataFrame()
for seed in range(20):
    pickle_filename = "F:\\PhD\\Studies\\NEN001\\Data\\Training_Out\\%s_pca_model.pkl" % seed
    with open(pickle_filename, 'rb') as file:
        pca_model = pickle.load(file)

    loadings = pca_model.components_

    df1=df1.append(pd.DataFrame(loadings))
    
plt.scatter(*loadings, alpha=0.3, label="Loadings");

plt.title("Loading plot");
plt.xlabel("Loadings on PC1");
plt.ylabel("Loadings on PC2");
plt.grid();
plt.legend(loc='lower left');
```

```python
export_csv = df1.to_csv ('..\\Data\\Output\\pca_loadings.csv', index = None, header=True)
```
